```python
from PIL import Image
import numpy as np
import albumentations as A

# Load image
img = np.array(Image.open("example.png"))

# Force apply augmentation (p=1.0 for demonstration)
aug = A.CoarseDropout(
    num_holes_range=(3, 6),
    hole_height_range=(10, 20),
    hole_width_range=(10, 20),
    fill="random_uniform",
    p=1.0  # Changed to always show effect
)

# Apply and display
result = aug(image=img)["image"]
Image.fromarray(result)
```
